# PC Games Sales Analysis

### Project Prequisites

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Preparing the Data

```python
df = pd.read_csv("Data\Games.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Question #1 Which genre is the most popular?

```python
plt.figure(figsize=(20,5))
sns.barplot(x = df["Genre"].head(175),y = df["Sales"])
plt.title("Ans: Battle Royale (as shown in the graph)")
plt.xticks(rotation = 90,size = 10)
plt.show()
```

### Question #2 Which publisher published most of the games?

```python
plt.figure(figsize=(25,5))
sns.countplot(x = df['Publisher'].head(137))
plt.title("Ans: Electronic Arts (as shown in the graph)")
plt.xticks(rotation = 90,size = 10)
plt.show()
```

### Question #3 Which developer developed most of the games?

```python
plt.figure(figsize=(25,5))
sns.countplot(x = df['Developer'].head(175))
plt.title("Ans:  Blizzard Entertainment (as shown in the graph)")
plt.xticks(rotation = 90,size = 10)
plt.show()
```

### Question #4 Which series is the most popular?

```python
plt.figure(figsize=(15,10))
sns.barplot(x = df["Series"].head(139),y = df["Sales"])
plt.title("Ans: Minecraft (as shown in the graph)")
plt.xticks(rotation=90,size=10)
plt.show()
```
